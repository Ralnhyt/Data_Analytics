## Métodos (I)
Un conductor va a una velocidad demasiado rápida, y un agente le detiene. Escribe una función con tres posibles resultados, *no ticket*, *small ticket* o *big ticket*, y que devuelva lo siguiente:

*  Si la velocidad es de 60 o menos, el resultado debe ser *no ticket*.
*  Si la velocidad está entre 61 y 80 (inclusive), el resultado debe ser *small ticket*.
*  Si la velocidad es superior a 80, el resultado debe ser *big ticket*.
*  Si es tu cumpleaños, puedes ir a la velocidad que quieras.

````python
def ticket(velocidad, cumpleaños):
    if cumpleaños:
        return "no ticket"
    elif  61 <= velocidad <= 81:
        return "no ticket"
    else: 
        return "big ticket"

print(ticket(50, False))
print(ticket(70, False))
print(ticket(90, False))
print(ticket(100, True))
    
````

# Avanzado: Clases (II)

Realizar una clase que administre una agenda. Se debe almacenar para cada contacto el nombre, el teléfono y el email. Además deberá mostrar un menú con las siguientes opciones:

*  Añadir contacto
*  Lista de contactos
*  Buscar contacto
*  Editar contacto
*  Cerrar agenda

````python
class contacto: 
    def __init__(self, nombre, telefono, email):
        self.nombre = nombre
        self.telefono = telefono
        self.email = email

    def __str__(self):
        return f"{self.nombre} - telefono: {self.telefono} - email: {self.email}"

class Agenda:
    def __init__(self):
        self.contactos = []
        
    def añadir_contacto(self):
        nombre = input("ingresa nombre: ")
        telefono = input("ingresa telefono: ")
        email = input("ingresa el email: ")
        self.contactos.append(contacto(nombre, telefono, email))
        print(f"contacto '{nombre}' añdido")

    def lista_contactos(self):
        if not self.contactos:
            print("No hay contactos.")
        else:
            for contacto in self.contactos:
                print(contacto)

    def buscar_contacto(self):
        nombre = input("ingresa nombre: ")
        for contacto in self.contactos:
            if contacto.nombre.lower() == nombre.lower():
                print("contacto encontrado:", contacto)
                return
        print("contacto no encontrado.")

    def editar_contacto(self):
        nombre = input("Ingrese nombre a editar: ")
        for contacto in self.contactos:
            if contacto.nombre.lower() == nombre.lower():
                contacto.telefono = input("nuevo telefono: ")
                contacto.email = input("nuevo email: ")
                print("contacto actualizado:", contacto)
                return
        print("contacto no encontrado.")

    def mostrar_menu(self):
        while True:
            print("\n1. añadir contacto")
            print("2. lista de contactos")
            print("3. buscar contacto")
            print("4. editar contacto")
            print("5. cerrar agenda")
            opcion = input("seleccione una opcon: ")

            if opcion == '1':
                self.añadir_contacto()
            elif opcion == '2':
                self.lista_contactos()
            elif opcion == '3':
                self.buscar_contacto()
            elif opcion == '4':
                self.editar_contacto()
            elif opcion == '5':
                print("cerrando agenda...")
                break
            else:
                print("opcion invalida.")

if __name__ == "__main__":  
    agenda = Agenda()
    agenda.mostrar_menu()
````
